````python
import pandas as pd

def import_csv(url):
  path = "https://drive.google.com/uc?export=download&id="+url.split("/")[-2]
  return pd.read_csv(path)

orderlines_url = "https://drive.google.com/file/d/1FYhN_2AzTBFuWcfHaRuKcuCE6CWXsWtG/view?usp=sharing"
orders_url = "https://drive.google.com/file/d/1Vu0q91qZw6lqhIqbjoXYvYAQTmVHh6uZ/view?usp=sharing"
products_url = "https://drive.google.com/file/d/1afxwDXfl-7cQ_qLwyDitfcCx3u7WMvkU/view?usp=sharing"
brands_url = "https://drive.google.com/file/d/1XGyabaa4mAkjixMk3XPgx_14OoSse3rs/view?usp=sharing"

orderlines_df = import_csv(orderlines_url)
orders_df = import_csv(orders_url)
product_df = import_csv(products_url)
brands_df = import_csv(brands_url)
````

#### Exercise 1: Convert the date column in the orderlines table into a datetime

#### Exercise 2: Use `.loc` to filter the orderlines data frame down to just rows where the `unit_price` does not have multiple periods. Then convert the column to numeric

````python
orderlines_df = orderlines_df.loc[~orderlines_df["unit_price"].str.contains(r'\.\d*\.')]
orderlines_df['unit_price'] =  orderlines_df['unit_price'].astype(float)
````

#### Exercise 3: Calculate the difference between the maximum and minimum `unit_price`

````python
minprice = orderlines_df["unit_price"].min()
maxprice = orderlines_df["unit_price"].max()
print("The difference between min ({:.2f}) and maxprice ({:.2f}) is {:.2f}".format(minprice, maxprice, maxprice - minprice))
````

#### Exercise 4: Get the total quantity of items grouped by the order and save it in a data frame called `order_quantities`

````python
order_quantities = orderlines_df.groupby("id_order")["product_quantity"].sum()
````

##### Hint:

`.groupby().agg({})`

#### Exercise 5: merge `order_quantities` with `orders_df`

````python
orders_df.merge(order_quantities,left_on="order_id", right_on="id_order")
````

````python
orders_df.merge(order_quantities,left_on="order_id", right_on="id_order").groupby('state')['product_quantity'].sum()
````
